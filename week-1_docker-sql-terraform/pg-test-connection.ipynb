```python
import pandas as pd
```

```python
df = pd.read_csv('green_tripdata_2019-09.csv', nrows=100)
```

```python
df
```

```python
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
```

```python
print(pd.io.sql.get_schema(df, name='green_tripdata_2019-09'))
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, name='green_tripdata_2019-09', con=engine))
```

```python
import time

start_time = time.time()

chunksize = 100000
df_iter = pd.read_csv('green_tripdata_2019-09.csv', iterator=True, chunksize=chunksize)

for df_chunk in df_iter:
    df_chunk.to_sql(name='green_tripdata_2019-09', con=engine, if_exists='append', index=False)

end_time = time.time()
execution_time = end_time - start_time

print(f"Execution time: {execution_time} seconds")
```
